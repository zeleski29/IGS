**16. Representation graphique de toutes les variables qualitatives**

```python
# conversion des types de donnees
df['sexe']=df['sexe'].astype('category')

# creation des listes de variables qualitatives et quantitatives
list_qual=df.select_dtypes(exclude='int64').columns
list_quant=df.select_dtypes(include=np.number).columns
```

```python
plt.figure(figsize=[17,6])# taille de la figure

plt.gcf().subplots_adjust( hspace = 0.5) 
#permet d'ajuster les positions des côtés des graphiques et les espacements entre ceux-ci
#hspace: espacement vertical entre deux graphiques

i=1
for var in list_qual :
    plt.subplot(3,3,i) # sectionner la grille en plusieurs parties
    x= df[var].value_counts()
    x.plot(kind='bar',)
    plt.title(var)
    i+=1
```

**17. Representation graphique de toutes les variables quantitative**

```python
plt.figure(figsize=[17,6])# taille de la figure

plt.gcf().subplots_adjust( hspace = 0.5)
#permet d'ajuster les positions des côtés des graphiques et les espacements entre ceux-ci
#hspace: espacement vertical entre deux graphiques

i=1
for var in list_quant :
    plt.subplot(2,3,i) # sectionner la grille en plusieurs parties
    plt.hist(df[var],edgecolor='white')
    plt.title(var)
    i+=1
```
